# Model Building

```python
from keras.models import Sequential
from keras.layers import Dense,Dropout

#Dense is responsible to create the layers(input,hidden,output)
#Dropout is responsible to drop the neurons randomly
```

```python
model = Sequential()  #This will stack the input layer one by one

#Hidden Layer_1
model.add(Dense(units=20,input_dim=8,activation="relu",kernel_initializer="glorot_uniform"))     #(8*20)+20 = 180
#Input layer is created along with the 1st Hidden layer...there is no specific code for creating input layer
#we have 8 features in dataset,so input_dim=8
#The Glorot uniform initializer,also called Xavier uniform initializer.Draws samples from a uniform distribution


#Hidden Layer_2
model.add(Dense(units=60,activation="relu",kernel_initializer="glorot_uniform")) #(20*60)+60 = 1260 parameters


#Hidden Layer_2
model.add(Dense(units=50,activation="relu",kernel_initializer="glorot_uniform")) #(60*50)+50 = 3050 parameters

#Output Layer
model.add(Dense(units=1, activation='sigmoid'))  #(50*1)+1 = 51

model.summary()
```

```python
import keras
```

```python
#If we directly put like rmsprop as optimizer it will function with its default value provided to it..
#we can also change(tune) that default value for our convinence by using below method

rms_opt =keras.optimizers.RMSprop(learning_rate=0.1,    #Here we tuned only the LR others are default value only
                                  rho=0.9,
                                  momentum=0.0,
                                  epsilon=1e-07,
                                  centered=False)
```

```python
model.compile(optimizer=rms_opt,loss="binary_crossentropy",metrics=["binary_accuracy"])

#Here Instead of (optimizer="rmsprop"),we change into (optimizer=rms_opt)like this to tune the rms default value
```

```python
model_train = model.fit(x=x_train,y=y_train,batch_size=40,epochs=10)   # (614/40)= 16 iterations
```

# Choose Best Value for Hyperparameters(Batch size,Epoch):

### To choose the Best hyperparameter value in ML we used  GridsearchCV which is available in sklearn library.To use GridsearchCV in DL,we need to make connection between sklearn and keras

```python
!pip install scikeras
```

```python
import numpy as np
import tensorflow as tf
from tensorflow import keras

from scikeras.wrappers import KerasClassifier
from sklearn.model_selection import GridSearchCV
```

```python
def create_model():      #It will create the model whenever it is being called

  model = Sequential()

  model.add(Dense(units=20,input_dim=8,activation="relu",kernel_initializer="glorot_uniform"))
  model.add(Dense(units=60,activation="relu",kernel_initializer="glorot_uniform"))
  model.add(Dense(units=50,activation="relu",kernel_initializer="glorot_uniform"))
  model.add(Dense(units=1, activation='sigmoid'))

  model.summary()

  model.compile(optimizer="adam",loss="binary_crossentropy",metrics=["binary_accuracy"])

  return model
```

```python
model = KerasClassifier(model=create_model, verbose=0) #Creating a Wrapper
```

```python
grid = GridSearchCV(estimator=model,param_grid={"batch_size":[10,20,30,40],"epochs":[10,15,20]})
```

# =======================PART-2==============================

```python
#Here we pass the values of hyperparameters(which are able to tune,independent of dataset) in class and call the class whenever needed.
#Note: Input and Output paramters are dependent on Dataset.so we cant tune that

def create_model(neuron1,neuron2,activation_function,init,learning_rate,dropout_rate):      #It will create the model whenever it is being called

  model = Sequential()

  #Hidden Layer_1
  model.add(Dense(units=neuron1,input_dim=8,activation=activation_function,kernel_initializer=init)) #Creating a variable as neuron1 to pass in the class later
  model.add(Dropout(dropout_rate))    #how many % we want to drop from hidden layer

  #Hidden Layer_2
  model.add(Dense(units=neuron2,activation=activation_function,kernel_initializer=init))
  model.add(Dropout(dropout_rate))

  #Output Layer
  model.add(Dense(units=1, activation='sigmoid',kernel_initializer="glorot_uniform"))

  model.summary()

  adam_tuned = keras.optimizers.Adam(learning_rate=learning_rate,
                                    beta_1=0.5,
                                    beta_2=0.96,
                                    epsilon=1e-07)

  model.compile(optimizer=adam_tuned,loss="binary_crossentropy",metrics=["binary_accuracy"])

  return model
```

```python
#neuron1 =[30,50,80]
#neuron2 =[60,60,90]
#activation_function =['relu','tanh']
#nit =['uniform','zero']
#learning_rate = [0.1,0.01,0.001]
#dropout_rate =[0.1,0.3]
#batch_size = [10,20]
#epochs = [10,12]

#By Using above,Create a Dictionary.so that it will be easy to pass those all directly
param_grid_dict = {
    'model__neuron1': [30, 50, 80],
    'model__neuron2': [60, 60, 90],
    'model__activation_function': ['relu', 'tanh'],
    'model__init': ['uniform', 'zero'],
    'model__learning_rate': [0.1, 0.01, 0.001],
    'model__dropout_rate': [0.1, 0.3],
    'batch_size': [10, 20],
    'epochs': [10, 12]
}
param_grid_dict
```

```python
model = KerasClassifier(build_fn = create_model)
```

```python
grid = GridSearchCV(estimator=model,param_grid = param_grid_dict)
```

```python
grid_results = grid.fit(scaled_x,y)
```

```python
grid_results.best_score_,grid_results.best_params_,grid_results.best_estimator_
```
