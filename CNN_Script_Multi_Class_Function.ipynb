```python
import pandas as pd
from tensorflow import keras
from tensorflow.keras.preprocessing.image import ImageDataGenerator
```

```python
train_data = ImageDataGenerator(rescale=1/255, shear_range=0.2, zoom_range=0.2, horizontal_flip=True, vertical_flip=True)

#Rescale=1/255(pixel range 0-255)...gives us the result btw 0 to 1
```

```python
test_data = ImageDataGenerator(rescale=1/255)
```

```python
train_data_generator = train_data.flow_from_directory(directory="/content/drive/MyDrive/Colab Notebooks/CNN/Train_Data",target_size=(256,256))
```

```python
test_data_generator = test_data.flow_from_directory(directory="/content/drive/MyDrive/Colab Notebooks/CNN/Test_Data")
```

```python
train_data_generator.image_shape
```

```python
from keras import Sequential
from keras.layers import Conv2D, MaxPooling2D, Flatten, Dense  #Dense is to build hidden layer
```

```python
model = Sequential()

#Convolutional_layer_1
model.add(Conv2D(input_shape=(256,256,3),filters=8, kernel_size=(3,3), strides= (1,1), padding="same",activation="relu",))
model.add(MaxPooling2D(pool_size=(2,2), strides=2))

#Convolutional_layer_2
model.add(Conv2D(filters=8, kernel_size=(3,3), strides= (1,1), padding="same",activation="relu"))  #Here we dont need input shape
model.add(MaxPooling2D(pool_size=(2,2), strides=2))

#Convolutional_layer_3
model.add(Conv2D(filters=16, kernel_size=(3,3), strides= (1,1), padding="same",activation="relu"))  #Here we dont need input shape
model.add(MaxPooling2D(pool_size=(2,2), strides=2))

#Flattening_layer
model.add(Flatten())       #(32*32*16)= we get 16384 rows

#Hidden_layer_1
model.add(Dense(units=30, activation="relu"))     #(16384*30)+30 = 491,550 parameters

#Output_layer
model.add(Dense(units=3, activation="softmax"))   #Image Classification problem..3 classes we have..so 3 neurons in output
                                                  #(30*3)+3= 93 parameters
model.summary()
```

```python
model.compile(optimizer="rmsprop",loss="categorical_crossentropy",metrics=["accuracy"])
```

```python
model = model.fit(x=train_data_generator,epochs=10,validation_data=test_data_generator) #Batch_size,we keep it default itself
```
