```python
#We are going to do some Data pre-processing to make our calculations faster.so for that we do Normalization of the Data

#Scaling(Divide with the pixel's Highest Number)
x_train = x_train/255
x_test = x_test/255

#so if x_train = 1 or 255 means....(1/255)=0.003 or (255/255)=1
#Anyways our x_train value is somewhere btw 0 to 255...so our value only ranges between 0 to 1 after Scaling.so our data get Normalized now.
```

# Steps:
# 1. Build the Architecture or Model (i.e Neural Network)
# 2. Model Compilation
# 3. Model Training
# 4. Model Testing

```python
from keras.models import Sequential #Sequential helps to stack layers on one Top of another
from keras.layers import Dense,Flatten    # Dense is responsible to build all layers which are input,Hidden,Output layers
```

```python
model = Sequential()

#Before creating  a input layer,we want to Flatten that
model.add(Flatten(input_shape=(28,28)))

model.add(Dense(units=50,activation="relu"))     #units=50 (no.of.neurons we want in hidden layer_1)
#we get output as 39250 parameters(784 inputs*50 hidden= 39200, 39200+50bias=39250 parameters)

model.add(Dense(units=10,activation="relu"))     #units=10 (no.of.neurons we want in hidden layer_2)
#we get output as 510 parameters(hidden_1*hidden_2)+ hidden_2 bias i.e (50*10)+10 = 510 parameters

model.add(Dense(units=10,activation="softmax"))  #For 10-class(Multi-class) Classification Problem,we use Softmax AF
#Based on the problem we have to decide the no.of.output units we should give(we have images btw o to 9,so there are 10 possible outputs)
#we get output as 110 parameters(hidden_2*output)+ output bias i.e (10*10)+10 = 110 parameters
model.summary()
```

```python
model.compile(optimizer="adam",loss="sparse_categorical_crossentropy",metrics=["sparse_categorical_accuracy"])

#Use **categorical_crossentropy**- Multi-class classification if the output lable is in One-Hot form
#Use **sparse_categorical_crossentropy**- Multi-class classification if the output lable is in integer form
```

```python
y_test    #See output is in the form of Integer
```

```python
model_training = model.fit(x = x_train,y = y_train,batch_size = 10,epochs=20,verbose= 'auto') #total images=60,000...so No.of.Iterations will be 6000 if batch size =10
```

```python
model_training = model.fit(x = x_train,y = y_train,batch_size = 32,epochs=5,verbose= 1,validation_data = (x_test,y_test))
```

```python
#in above result we get similar results for both Training data =99% and Test data = 97% moreorless, so our model is not OverFitting
```

```python
y_pred_test = model.predict(x_test)
```
