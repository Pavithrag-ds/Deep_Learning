```python
from tensorflow import keras
from keras.datasets import mnist
from keras.models import Sequential
from keras.layers import Conv2D,AveragePooling2D,Flatten,Dense
import numpy as np
```

```python
(x_train, y_train), (x_test, y_test) = keras.datasets.mnist.load_data(path="mnist.npz")
```

```python
x_train.shape     #our MNIST datset have 28x28 but our Lenet architecture needs 32x32
```

```python
x_train = np.pad(array=x_train,pad_width=((0,0),(2,2),(2,2)))     #Padding operation to make sure our data is (32x32x1)by increasing the width of the padding
x_test = np.pad(array=x_test,pad_width=((0,0),(2,2),(2,2)))
```

```python
x_train.shape, x_test.shape #if we put pad_width=((0,0),(1,1),(1,1), we will get (60000, 31, 31)as output
```

```python
import pandas as pd
pd.DataFrame(x_train[3000])    #In output,we can see 9 in the dataframe(Which is the 3000th element present in our dataset)
```

```python
y_train[3000]
```

```python
x_train = x_train/255
x_test = x_test/255
```

```python
pd.DataFrame(x_train[3000])    #In this output also we get 9 but it is not visible to our eyes(the values forms like 9)
```

```python
model = Sequential()

#Convolutional_Layer_1
model.add(Conv2D(input_shape=(32,32,1),filters=6, kernel_size=(5,5), strides=(1, 1), activation='tanh'))
model.add(AveragePooling2D(pool_size=(2,2), strides=(2, 2)))

#Convolutional_Layer_2
model.add(Conv2D(filters=16, kernel_size=(5,5), strides=(1, 1), activation='tanh'))
model.add(AveragePooling2D(pool_size=(2,2), strides=(2, 2)))

#Convolutional_Layer_3
model.add(Conv2D(filters=120, kernel_size=(5,5), strides=(1, 1), activation='tanh'))

#Flatten_Layer
model.add(Flatten())

#Hidden_Layer
model.add(Dense(units=84, activation='tanh'))

#Output_Layer
model.add(Dense(units=10, activation='softmax',))    #(0-9)digit class

model.summary()
```

```python
#In above output,we get 5 trainable parameters..so we called this as **LeNet-5**
```

```python
model.compile(optimizer='rmsprop',loss="sparse_categorical_crossentropy",metrics=["sparse_categorical_accuracy"])

#Whenever we have integers as output,we have to use "sparse_categorical_crossentropy"
```

```python
model_train=model.fit(x=x_train,y=y_train,epochs=15,validation_data=(x_test,y_test))
```

```python
model.evaluate(x_test,y_test)
```

```python
predict = model.predict(x_test)
```
